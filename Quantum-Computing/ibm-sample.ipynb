```python
from qiskit import QuantumCircuit
from qiskit_aer.primitives import SamplerV2  # Aer implementation

# Build a simple Bell circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

# Use local Aer-based SamplerV2
sampler = SamplerV2()  # default_shots=1024
job = sampler.run([qc])   # PUB list: [circuit1, circuit2, ...]
result = job.result()

print("Result object:", result)
print("First PUB quasi distribution:", result)
```
